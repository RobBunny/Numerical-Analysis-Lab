```python
#Newton-Raphson method of finding roots
import numpy as np
import matplotlib.pyplot as plt
import math as m

#Newton-Raphson function
def new(x):
  g = 9.81
  y = 36
  c = 15
  t = 10
  A = (g * x) / c
  B = 1 - m.exp(-(c/x) * t)
  result = A * B - y
  return result

def new_der(x):
  g = 9.81
  c = 15
  t = 10
  B = m.exp(-(c/x) * t)
  result = (g / c) * (1 - B - (c*t / x) * B)
  return result

#Given data
es = 0.001
ea = 100
i = 0
mk_old = 40

#Plotting Criterion
iterations = []
approx_errors = []

#Root checking logic
print(f"{'Iter':<5} {'mk':<15} {'f(mk)':<20} {'f\'(mk)':<20} {'ea':<15}")
while ea > es:
  i += 1
  mk = mk_old - (new(mk_old) / new_der(mk_old))

  ea = abs((mk - mk_old)/mk) * 100
  print(f"{i:<5} {mk:<15.10f} {new(mk):<20.10f} {new_der(mk):<20.10f} {ea:<15.10f}")

  iterations.append(i)
  approx_errors.append(ea)

  mk_old = mk

print(f"\nThe root is {mk:.10f} (Approx.)")

#Graph Plot
plt.figure(figsize=(8,5))
plt.plot(iterations, approx_errors, marker='o')
plt.title('Approximate Relative Error vs Iteration')
plt.xlabel('Iteration')
plt.ylabel('Approximate Relative Error (%)')
plt.grid(True)
plt.show()
```
